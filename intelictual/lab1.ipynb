Задание 1:

Переписать функцию Sine , таким образом, чтобы на выходе получать сигнал косинуса вместо синуса (np.cos)

```python
import matplotlib.pyplot as plt
import numpy as np

def Cosine(A, f, phi, fs, t, N):
  T = t/fs
  n =np.arange(N)
  x = A*np.cos(2*np.pi*f*n*T + phi)
  return x, n

A = 1.0
f = 10.0
phi = 1.0
fs = 50.0
t = 0.1


plt.plot(Cosine(A,f,phi, fs, t, 100) [0])
plt.title ('Cosine wave')
plt.xlabel ('Time')
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.show()
```

Задание 2:

Модифицировать функцию Sine таким образом чтобы можно было взвращать составной сигнал - состоящий из комбинаций sin и cos, каждая отдельная гармоническая функция в составном сигнале должна принимать свои параметры:

A = Амплитуда сигнала

f = Частота сигнала Hz

phi = Начальная фаза синусоиды "Радианы"

Сделайте вариант с 2-мя и 3-мя гармоническими сигналами

```python
def DoubleSignal(sin_A, sin_f, sin_phi, cos_A, cos_f, cos_phi, fs, t, N):
  T = t/fs
  n =np.arange(N)
  sin = sin_A*np.sin(2*np.pi*sin_f*n*T + sin_phi)
  cos = cos_A*np.cos(2*np.pi*cos_f*n*T + cos_phi)
  signal = sin+cos
  return signal, n

def TripleSignal(sin_A, sin_f, sin_phi, cos_A, cos_f, cos_phi,
                      sin2_A, sin2_f, sin2_phi, fs, t, N):
  T = t/fs
  n =np.arange(N)
  sin = sin_A*np.sin(2*np.pi*sin_f*n*T + sin_phi)
  cos = cos_A*np.cos(2*np.pi*cos_f*n*T + cos_phi)
  sin2 = sin2_A*np.sin(2*np.pi*sin2_f*n*T + sin2_phi)
  signal = sin+cos+sin2
  return signal, n

sin_A, cos_A, sin2_A = 1.0, 0.5, 0.7
sin_f, cos_f, sin2_f = 22050, 4400, 5000
sin_phi, cos_phi, sin2_phi = 1.0, 0, 1.2
fs = 50.0
t = 0.1

wave, n = DoubleSignal(sin_A, sin_f, sin_phi, cos_A, cos_f, cos_phi, fs, t, 100)

plt.plot(wave)
plt.title('Composite signal sin and cos')
plt.xlabel("Time")
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.show()

wave2, n2 = TripleSignal(sin_A, sin_f, sin_phi, cos_A, cos_f, cos_phi,
                      sin2_A, sin2_f, sin2_phi, fs, t, 100)

plt.plot(wave2)
plt.title('Composite signal sin, cos and sin')
plt.xlabel("Time")
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.show()
```

Сигнал задан следующей функцией: t = np.linspace(0, 1, n, endpoint=True) np.sin(3*np.pi*t) + np.sin(5*np.pi*t) Отобразить на графике Аналоговый, Дискретный и Квантованные сигналы

```python
n = 40
t = np.linspace(0,1,n,endpoint=True)
x = np.sin(3*np.pi*t)+np.sin(5*np.pi*t)

def plt_sel(s, *args, **kwargs):
    if s == 0:
        return plt.plot(*args, **kwargs)
    if s == 1:
        return plt.stem(*args)
    if s == 2:
        return plt.step(*args, **kwargs)

t_titles = ['Аналоговый', 'Дискретный', 'Квантованный']

fig = plt.figure(figsize=(16, 4), dpi=80)
for i in range(3):
  plt.subplot(1, 3, i+1)
  plt.title(t_titles[i])
  plt_sel(i, t, x)
  plt.xlim([0, 1])
  plt.yticks(np.linspace(np.floor(np.min(x)), np.ceil(np.max(x)), 9))
  plt.grid(True)

  ax = plt.gca()
  ax.spines['right'].set_color('none')
  ax.spines['top'].set_color('none')
  ax.spines['bottom'].set_position(('data', 0))

plt.tight_layout()

```

Дискре́тный сигна́л — сигнал, который является прерывистым и который изменяется во времени и принимает любое значение из списка возможных значений. Квантовый — в обработке сигналов — разбиение диапазона отсчётных значений сигнала на конечное число уровней и округление этих значений до одного из двух ближайших к ним уровней.

Задача 3(PRO):

Самостоятельно ознакомившись с тем что такое Автокорреляционная функция рассчитать и отобразить графически Автокорреляционную функцию прямоугольного импульса: [np.zeros(8), np.ones(3), np.zeros(8)] , Напишите функцию вычисляющую автокорреляцию.

```python
def Autocorrelation(x):
    N = len(x)
    result = np.zeros(2*N-1)
    for k in range(-N+1, N):
        sum_val = 0
        for n in range(max(0, -k), min(N, N-k)):
            sum_val += x[n] * x[n + k]
        result[k + N - 1] = sum_val
    return result

imp = np.concatenate([np.zeros(8), np.ones(3), np.zeros(8)])
acf = Autocorrelation(imp)

plt.figure(figsize=(10, 4))
plt.stem(np.arange(len(acf)), acf)
plt.title('Автокорреляционная функция прямоугольного импульса')
plt.xlabel('Лаг')
plt.ylabel('Автокорреляция')
plt.grid(True)
plt.show()
```

```python
!pip install neurokit2
from scipy.signal import spectrogram
import neurokit2 as nk

ecg = nk.ecg_simulate(duration=10, sampling_rate=100, heart_rate=70)
t = np.linspace(0, 10, len(ecg))

N = len(ecg)
X = fft(ecg)
freqs = np.fft.fftfreq(N, d=t[1]-t[0])
magnitude = np.abs(X) / N * 2
magnitude[0] /= 2

f, t_spec, Sxx = spectrogram(ecg, fs=100)

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 10))

ax1.plot(t, ecg)
ax1.set_title('ЭКГ сигнал')
ax1.set_xlabel('Время')
ax1.set_ylabel('Амплитуда')

ax2.plot(freqs[:N//2], magnitude[:N//2])
ax2.set_title('Амплитудный спектр')
ax2.set_xlabel('Частота')
ax2.set_ylabel('Амплитуда')

im = ax3.pcolormesh(t_spec, f, 10*np.log10(Sxx), shading='gouraud')
ax3.set_title('Спектрограмма')
ax3.set_xlabel('Время')
ax3.set_ylabel('Частота')
plt.colorbar(im, ax=ax3)

plt.tight_layout()
plt.show()

```
